```python
import json,pandas,glob #Appending all the products info from multiple JSON files to single dataframe
import os.path

d1 = []
d2 = []
read_files = glob.glob("AmazonReviews/laptops/*.json")
for i in range(0,len(read_files)):
    with open(read_files[i], 'rb') as x:
        JSONFILE = json.load(x)
    prod = JSONFILE['ProductInfo']
    Review = JSONFILE['Reviews']
    d1.append(prod)
    d2.append(Review)
Products = pandas.DataFrame(d1)
Reviews = pandas.DataFrame(d2)

merge = pandas.concat([Reviews, Products.ProductID], axis=1,join = 'inner')

merge
```

```python
Prod_Json = Products.to_json(orient = 'records') #writing all the Product records to a seperate JSON for ease of access

with open("ProductInfo.json", "w") as f:
    f.write(Prod_Json)
```

```python
Reviews_Json = Reviews.to_json(orient = 'records') #writing all the Review records to a seperate JSON for ease of access

with open("Reviews.json", "w") as f:
    rew = f.write(Reviews_Json)
```

```python
with open("merged_file.json", "wb") as outfile: #saving all JSON files to a single JSON file
    outfile.write('[{}]'.format(
        ','.join([open(f, "rb").read() for f in read_files])))

with open("merged_file.json") as fs:
    FL = json.load(fs)
    
pandas.DataFrame(FL , columns = ['Reviews'])
```
